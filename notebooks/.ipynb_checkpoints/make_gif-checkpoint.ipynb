# Generating animated timelines of salamander hand development under normal and abnormal conditions.

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from aicsimageio import imread
from skimage.transform import resize
```

```python
# load image series
samples = ["normal", "abnormal"]
image_shape = (512, 512)
series = {}
times = {}
path = Path("../data/time-series_normal&abnormal/")
for sample in samples:
    print(f"Load path: {path / sample}")
    series[sample] = []
    times[sample] = []
    for file in (path / sample).glob("*.png"):
        # resize image
        img = imread(file)
        img = img[0][0][0]
        img = np.mean(img, axis=2)
        img = resize(img, output_shape=image_shape)
        series[sample].append(img)
        times[sample].append(file.stem)
    # sort series and times, according to times order
    # the sort key is the time point name strip the postfix 'W'
    series[sample] = np.array(series[sample])
    times[sample] = np.array([int(t.rstrip('W')) for t in times[sample]])
    sort_idx = np.argsort(times[sample])
    series[sample] = series[sample][sort_idx]
    times[sample] = times[sample][sort_idx]
    print(f"Time points: {times[sample]}")
```

```python
# plot image series
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.set_title('normal')
ax2.set_title('abnormal')
img1 = ax1.imshow(series["normal"][0], cmap="gray")
img2 = ax2.imshow(series["abnormal"][0], cmap="gray")

def update(frame):
    img1.set_data(series["normal"][frame])
    img2.set_data(series["abnormal"][frame])

ani = FuncAnimation(
    fig, update, frames=len(series["normal"]),
    interval=300)
ani.save('../figs/normal&abnormal.gif', writer='imagemagick', fps=4)
```
